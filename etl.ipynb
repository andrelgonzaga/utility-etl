```python


class FileHandler():
    
    @staticmethod
    def get_file_names(dir_path):
        return os.listdir(dir_path)
    
    @staticmethod
    def get_excel_names(dir_path):
        file_names = FileHandler.get_excel_names(dir_path)
        excel_names = []
        for file_name in file_names:
            if('.xlsx' in file_name):
                excel_names.append(file_name)
        return excel_names
    
    @staticmethod
    def get_excel_paths(dir_path):
        pass
    
    @staticmethod
    def load_df_from_excel(excel_name, sheet_name, header_row):
        df = pd.read_excel(excel_name, sheet_name=sheet_name, header=header_row-1)
        return df
    
    @staticmethod
    def save_excel_from_df(df, excel_name, sheet_name, header_row):
        wb = openpyxl.load_workbook(filename=excel_name)
        ws = wb[sheet_name]
        header = df.columns.tolist()
        rows_count = df.shape[0] 
        cols_count = df.shape[1]
        for row in range(header_row, rows_count + header_row + 1):
            for col in range(1, cols_count + 1):
                if(row == header_row):
                    ws.cell(row=row, column=col).value = header[col - 1]
                else:
                    ws.cell(row=row, column=col).value = df.iat[row - 2, col - 1]
        ws.save(excel_name)
        
    @staticmethod
    def clear_excel_keeping_format(excel_name, sheet_name, header_row):
        wb = openpyxl.load_workbook(filename=excel_name)
        ws = wb[sheet_name]
        rows_count = ws.max_row
        cols_count = ws.max_column
        for row in range(header_row + 1, rows_count + 1):
            for col in range(1, cols_count + 1):
                ws.cell(row=row, column=col).value = None
        wb.save(excel_name)
    
    @staticmethod
    def get_excel_header(excel_name, sheet_name, header_row):
        wb = openpyxl.load_df_into_excel(filename=excel_name)
        ws = wb[sheet_name]
        header = []
        cols_count = ws.max_column
        for col in range(1, cols_count + 1):
            header.append(ws.cell(row=header_row, column=col).value)
        return header
    
    @staticmethod
    def get_df_header(df):
        return df.columns.tolist()
    
    @staticmethod
    def get_excel_cols_count(excel_name, sheet_name, header_row):
        header = FileHandler.get_excel_header(excel_name, sheet_name, header_row)
        return len(header)
    
    @staticmethod
    def delete_file(file_path):
        os.remove(file_path)
        
    @staticmethod
    def delete_files(file_paths):
        for file_path in file_paths:
            os.remove(file_path)
```
